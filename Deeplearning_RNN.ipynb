```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
### Vocabulary size
voc_size=500
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
## Pre padding done to make sure sentence size are of equal
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# Feature represenation size or dimension is 10
dim=10
```

```python
# To fetch first sentence from the dataset - represented in the form of index format
embedded_docs[0]
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
# The prediction done for the first statement
model.predict(embedded_docs[0])
```

```python
# The prediction done for all the statements
print(model.predict(embedded_docs))
```
